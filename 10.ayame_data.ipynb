```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
def pie_chart(any_graph):
    df_sample = any_graph.value_counts()
    plt.pie(df_sample,autopct="%1.1f%%")
    plt.show()
    
# 回答
# 1.　ラベル毎の円グラフを表示
print("--------1. 円グラフの表示--------")
print(pie_chart(df["sepal_length"]))
print(pie_chart(df["sepal_width"]))
print(pie_chart(df["petal_length"]))
print(pie_chart(df["petal_width"]))
print(pie_chart(df["Species"]))
```

## 【問題7】特徴量間の関係の確認

特徴量間の関係を見ていきます。次の4つを行ってください。


**1.４つの特徴量から２つ選ぶようなすべての組み合わせ（６つ）の散布図を種類ごとに色分けして作る。**

以下の図はそのうち一つの例です。





**2.すべての特徴量の組み合わせを一度に表示させた、散布図行列を作る。**



seaborn.pairplot - seaborn 0.11.1 documentation


**3.　４つの特徴量に関する相関係数行列を作る**


相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。


pandas.DataFrame.corr — pandas 0.24.2 documentation


**4.　相関係数行列のヒートマップを作る(１つ)**

seaborn.heatmap - seaborn 0.11.1 documentation

```python
import seaborn as sns
# データセットの各列を変数に代入しておく
sepal_length = df.loc[:, ["sepal_length"]]
sepal_width = df.loc[:, ["sepal_width"]]
petal_length = df.loc[:, ["petal_length"]]
petal_width= df.loc[:, ["petal_width"]]
species= df.loc[:, ["Species"]]
```

```python
# 1.４つの特徴量から２つ選ぶようなすべての組み合わせ（６つ）の散布図を種類ごとに色分けして作る。
# c = linspace(1,10,length(x));
colorlist = ['#e41a1c', '#377eb8', '#4daf4a']

# ①
plt.title("①")
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.scatter(sepal_length,sepal_width,color=colorlist)
plt.show()

# ②
plt.title("②")
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
plt.scatter(sepal_length,petal_length,color=colorlist)
plt.show()

# ③
plt.title("③")
plt.xlabel("sepal_length")
plt.ylabel("petal_width")
plt.scatter(sepal_length,petal_width,color=colorlist)
plt.show()

# ④
plt.title("④")
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
plt.scatter(sepal_width,petal_length,color=colorlist)
plt.show()

# ⑤
plt.title("⑤")
plt.xlabel("sepal_width")
plt.ylabel("petal_width")
plt.scatter(sepal_width,petal_width,color=colorlist)
plt.show()

# ⑥
plt.title("⑥")
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.scatter(petal_length,petal_width,color=colorlist)
plt.show()

```

```python
# 2.すべての特徴量の組み合わせを一度に表示させた、散布図行列を作る。
sns.pairplot(df,  hue="Species")
# Setosa ... ０  (blue)
# Versicolor　...1  (orange)
#  Virginica...2  (green)
```

```python
# 3.　４つの特徴量に関する相関係数行列を作る

# 相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。
corr_mat = df.corr()
print(corr_mat)


# 4.　相関係数行列のヒートマップを作る(１つ)
sns.heatmap(corr_mat,
                        center=0,
                        annot=True,
                       )
 # True:格子の中に値を表示
#  fmt='.1f', 小数点何位まで表示するか指定
plt.show()
```
