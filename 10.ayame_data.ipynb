```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
def pie_chart(any_graph):
    df_sample = any_graph.value_counts()
    plt.pie(df_sample,autopct="%1.1f%%")
    plt.show()
    
# 
# 1.
print("--------1. --------")
print(pie_chart(df["sepal_length"]))
print(pie_chart(df["sepal_width"]))
print(pie_chart(df["petal_length"]))
print(pie_chart(df["petal_width"]))
print(pie_chart(df["Species"]))
```
